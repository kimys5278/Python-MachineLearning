```python

# 입력할 오디오 파일 경로와 STFT 설정값
audio_path = './train/TRAIN_0000.wav'
fft = 2048
hl = 512

# 오디오 파일 불러오기
audio, sr = librosa.load(audio_path)

# STFT 결과 계산
stft = np.abs(librosa.stft(audio, n_fft=fft, hop_length=hl))

# STFT 결과 그래프 그리기
plt.figure(figsize=(10, 5))
plt.title("STFT Magnitude")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.imshow(np.log10(stft), origin="lower", cmap="jet", aspect="auto", extent=[0, len(audio) / sr, 0, sr / 2])
plt.colorbar()
plt.show()

```

### 위 코드로 그려진 그래프는 시간-주파수 평면에서 STFT 결과의 크기(magnitude)를 나타낸 것.

### x축은 시간(time)을 나타내며, 단위는 초(s)입니다. y축은 주파수(frequency)를 나타내며, 단위는 헤르츠(Hz)입니다. 그래프의 색상은 STFT 결과의 크기를 로그 스케일로 변환한 값을 적용한 것.

### 로그 스케일은 큰 값을 좁은 범위에 매핑하여 작은 차이를 더 잘 볼 수 있게 해줍니다. 이러한 로그 스케일로 변환한 STFT 결과값이 픽셀 값으로 그래프에 표시됨.

### 그래프의 좌측 하단은 주파수의 최소값(0 Hz)과 시간의 최소값(0 s)을 나타내며, 우측 상단은 주파수의 최대값(샘플링 주파수의 절반)과 시간의 최대값(오디오 파일의 길이)을 나타냅.

### 색상 바(colorbar)는 STFT 결과값에 대응하는 크기를 나타내며, 그래프 상에서 색상이 짙을수록 큰 값을 갖음.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# 오디오 파일 불러오기
audio, sr = librosa.load('./train/TRAIN_0002.wav')

# STFT 설정값
n_fft = 2048
hop_length = 512

# Mel 스펙트로그램 계산
S = librosa.feature.melspectrogram(y=audio, sr=sr, n_fft=n_fft, hop_length=hop_length)

# Mel 스펙트로그램 그래프 그리기
plt.figure(figsize=(10, 5))
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

```python

import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# 입력할 오디오 파일 경로와 STFT 설정값
# 파일 경로와 라벨
file_paths = ['./train/TRAIN_0006.wav','./train/TRAIN_0000.wav','./train/TRAIN_0001.wav','./train/TRAIN_0005.wav','./train/TRAIN_0002.wav','./train/TRAIN_0003.wav']
labels = ['label_1', 'label_2', 'label_3', 'label_4', 'label_5', 'label_6']

# 파라미터 설정
fft = 2048
hop_length = 512

# STFT 결과를 저장할 리스트
stfts = []

# 파일별로 STFT를 계산하고 리스트에 저장하기
for file_path in file_paths:
    audio, sr = librosa.load(file_path)
    stft = np.abs(librosa.stft(audio, n_fft=fft, hop_length=hop_length))
    stfts.append(stft)

# 그래프 옵션 설정
plt.figure(figsize=(15, 10))
plt.subplots_adjust(hspace=0.4)

# 리스트에 저장된 STFT 결과로 그래프 그리기
for i, stft in enumerate(stfts):
    plt.subplot(3, 2, i+1)
    librosa.display.specshow(librosa.amplitude_to_db(stft, ref=np.max), sr=sr, y_axis='log', x_axis='time')
    plt.title(labels[i])
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()

plt.show()

```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# wav 파일 경로 및 라벨 정의
file_paths = ['./train/TRAIN_0006.wav','./train/TRAIN_0000.wav','./train/TRAIN_0001.wav','./train/TRAIN_0005.wav','./train/TRAIN_0002.wav','./train/TRAIN_0003.wav']
# STFT 결과 계산
fft = 2048
hl = 512
stfts = []
for path in file_paths:
    audio, sr = librosa.load(path)
    stft = np.abs(librosa.stft(audio, n_fft=fft, hop_length=hl))
    stfts.append(stft)

# 그래프 그리기
plt.figure(figsize=(15, 10))
plt.title("STFT Magnitude Comparison")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")

colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow']
for i in range(len(stfts)):
    plt.plot(np.log10(stfts[i]), color=colors[i], label=labels[i])

plt.legend()
plt.show()
```
